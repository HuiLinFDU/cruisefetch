### Python Script for Capstone Project
TASK : Statistics on the running time of the simulation in the log information

```python
import os
import csv

# Define the path to the directory containing the log files
log_directory = '/content/sample_data/logfile'  # Update this path to your actual log directory

# Define the path to the output CSV file
output_csv = '/content/sample_data/output.csv'  # Update this path to your desired output location

# Initialize counters
total_files = 0
matched_lines = 0  # Initialize matched_lines counter

# Open the CSV file for writing
with open(output_csv, 'w', newline='') as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)

    # Write the header row
    csvwriter.writerow(['Filename', 'Simulation Time'])

    # Iterate over all files in the specified directory
    for filename in os.listdir(log_directory):
        # Check if the file ends with '.out'
        if filename.endswith('.out'):
            total_files += 1  # Increment total file count

            # Construct the full file path
            file_path = os.path.join(log_directory, filename)

            # Open and read the file
            with open(file_path, 'r') as file:
                # Iterate over each line in the file
                for line in file:
                    # Check if the line contains the specific string
                    if "Finished CPU 0 instructions:" in line:
                        # Find the start and end positions of the desired substring
                        start = line.find("(Simulation time:")
                        end = line.find(")", start)

                        # Extract the substring if found
                        if start != -1 and end != -1:
                            simulation_time = line[start:end+1]

                            # Write the filename and simulation time to the CSV file
                            csvwriter.writerow([filename, simulation_time])

                            # Increment matched lines counter
                            matched_lines += 1

                            # Print a confirmation message (optional)
                            print(f"Processed file: {filename}, Simulation Time: {simulation_time}")
        else:
            # Print a message if the file is not a .out file
            print(f"Skipping non-log file: {filename}")

# Print statistics
print("\n--- Statistics ---")
print(f"Total files processed: {total_files}")
print(f"Total matched lines: {matched_lines}")
print(f"All data has been saved to {output_csv}")
```
